```python
df2=df[(df['Event distance/length'].isin(['50km','50mi'])) & (df['Event distance/length'].isin(['50km','50mi'])) & (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0)=='USA')]
```

```python
df2.shape
```

```python
#remove USA from event name
```

```python
df2['Event name']=df2['Event name'].str.split('(').str.get(0)
```

```python
df2.head()
```

```python
#clean up athlete age
df2.dtypes
```

```python
df2.dropna()
```

```python
df2['Athlete year of birth']=df2['Athlete year of birth'].astype(float)
```

```python
df2['athlete age']=2020-df2['Athlete year of birth']
```

```python
df2['Athlete performance']=df2['Athlete performance'].str.split(' ').str.get(0)
```

```python
df2.head(5)
```

```python
#drop columns: club, country, year of birth, age category
```

```python
df2=df2.drop(['Athlete club','Athlete country','Athlete year of birth','Athlete age category'],axis=1)
```

```python
df2.head(5)
```

```python
df2[df2['athlete age'].isna()==1]
```

```python
df2=df2.dropna()
df2
```

```python
df2.shape
```

```python
#check for dupes
```

```python
df2[df2.duplicated()==True]
```

```python
#reset index
```

```python
df2.reset_index(drop=True)
```

```python
df2.dtypes
```

```python
df2['athlete age']=df2['athlete age'].astype(int)
```

```python
df2.dtypes
```

```python
#fix types
df2['Athlete average speed']=df2['Athlete average speed'].astype(float)
```

```python
df2=df2.rename(columns={'Year of event':'year',
                        'Event dates':'race_day',
                        'Event name':'race_name',
                        'Event distance/length':'race_length',
                        'Event number of finishers':'race_number_of_finishers',
                        'Athlete performance':'athlete_performance',
                        'Athlete gender':'athlete_gender',
                        'Athlete average speed':'athlete_average_speed',
                        'Athlete ID': 'athlete_id'
                       })
```

```python
df2.head()
```

```python
#reorder columns
```

```python
df3=df2[['race_day','race_name','race_length','race_number_of_finishers','athlete_performance','athlete_gender','athlete_average_speed','athlete_id','athlete age']]
```

```python
df3.query('race_length=="50mi"').groupby('athlete age')['athlete_average_speed'].agg(['mean','count']).sort_values('mean',ascending=False).query('count>19')
```

```python
df3.query('race_length=="50mi"').groupby('athlete age')['athlete_average_speed'].agg(['mean','count']).sort_values('mean',ascending=True).query('count<9')
```
